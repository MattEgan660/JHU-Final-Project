```python
import pandas as pd 
import math 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import numpy as np
import datetime as dt
%matplotlib inline
```

```python
df1 = pd.read_csv('csv/nvidia.csv')
df1.head()
```

```python
df1.info()
```

```python
df1['date'] = pd.to_datetime(df1['date'])
df1.set_index('date', inplace = True)
df1.head()
```

```python
df1.info()
```

```python
days = (df1.index[-1] - df1.index[0]).days
cagr = ((((df1['adjClose'][-1] /  df1['adjClose'][1])) ** (365/days)) -1)
```

```python
df1['returns'] = df1['adjClose'].pct_change()  
volatility = df1['returns'].std()*math.sqrt(252)
```

```python
print('CAGR=',str(round(cagr,4)*100)+"%")
print('Annual Volatility =',str(round(volatility,4)*100)+"%")
```

```python
plt.figure(figsize = (10,6))
result = []


#Define Variables 
S  = df1['adjClose'][-1] #Starting stock which is the last know stock price. 
T = 252 #Number of trading days in the year 
Mu = cagr
vol = volatility 
iterations = 1000 #Number of simulations


for i in range(iterations): 
    daily_returns = np.random.normal(Mu/T, vol/math.sqrt(T), T) + 1
    price_list = [S]
    
    for x in daily_returns:
        price_list.append(price_list[-1]*x)
        
        plt.plot(price_list)
        result.append(price_list[-1])

plt.show()

#histogram to show distribution of results
plt.hist(result, bins=100)
plt.show()
```
